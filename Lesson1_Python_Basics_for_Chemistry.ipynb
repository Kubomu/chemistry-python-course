## Example 1: pH Calculation

### The Chemistry:
pH is a measure of acidity/basicity, calculated using:

**pH = -log₁₀[H⁺]**

Where [H⁺] is the hydrogen ion concentration in mol/L

### Real-World Context:
NEMA (National Environment Management Authority) requires bottling companies in Uganda to maintain water pH between 6.5 and 8.5.

Let's calculate pH values for quality control!

```python
# Import the math library for logarithm calculations
import math

# Example 1a: Calculate pH of a solution with known H+ concentration
# Crown Beverages water sample
h_concentration = 0.00001  # This is 1.0 x 10^-5 mol/L (scientific notation: 1e-5)

# Calculate pH using the formula: pH = -log10(H+)
pH_value = -math.log10(h_concentration)

# Display the result
print(f"H⁺ concentration: {h_concentration} mol/L")
print(f"Calculated pH: {pH_value}")
print(f"pH (rounded): {round(pH_value, 2)}")

# Check if it meets NEMA standards
if 6.5 <= pH_value <= 8.5:
    print("✅ PASS: Water meets NEMA pH standards")
else:
    print("❌ FAIL: Water does NOT meet NEMA pH standards")
```

```python
import math

# Example 1b: Interactive pH Calculator
# You can input your own H+ concentration!


print("=== Interactive pH Calculator ===")
print("Enter H⁺ concentration in scientific notation")
print("Example: For 1.0 x 10^-7, enter 0.0000001 or 1e-7\n")

# Get input from user (you!)
h_input = float(input("Enter H⁺ concentration (mol/L): "))

# Safety check: H+ concentration must be positive
if h_input <= 0:
    print("Error: H⁺ concentration must be greater than 0")
else:
    # Calculate pH
    calculated_pH = -math.log10(h_input)
    
    # Show result with interpretation
    print(f"\nCalculated pH: {round(calculated_pH, 2)}")
    
    # Classify the solution
    if calculated_pH < 7:
        print("🔴 This solution is ACIDIC")
    elif calculated_pH == 7:
        print("🟢 This solution is NEUTRAL")
    else:
        print("🔵 This solution is BASIC/ALKALINE")
```

- **Print statements**: Displaying results with `print()`
- **f-strings**: Embedding variables in text with `f"{variable}"`

## Example 2: Molar Mass Calculator

### The Chemistry:
Molar mass is the mass of one mole of a substance (g/mol). It's calculated by adding up the atomic masses of all atoms in a molecule.

**Example:** Water (H₂O) = 2(H) + 1(O) = 2(1.008) + 16.00 = 18.016 g/mol

### Real-World Context:
Pharmaceutical companies in Kampala need accurate molar mass calculations for drug formulation and quality control.

```python
# Store atomic masses in a dictionary (like a chemistry reference table)
# Dictionary format: "element_symbol": atomic_mass
atomic_masses = {
    "H": 1.008,   # Hydrogen
    "C": 12.011,  # Carbon
    "N": 14.007,  # Nitrogen
    "O": 16.00,   # Oxygen
    "S": 32.06,   # Sulfur
    "Cl": 35.45,  # Chlorine
    "Na": 22.99,  # Sodium
    "Ca": 40.08   # Calcium
}

print("Available elements:", list(atomic_masses.keys()))
print("\nExample: Atomic mass of Carbon (C):", atomic_masses["C"], "g/mol")
print("\nExample: Atomic mass of Nitrogen (N):", atomic_masses["N"], "g/mol")
```

```python
# Example 2a: Calculate molar mass of Water (H₂O)
print("=== Calculating Molar Mass of Water (H₂O) ===")

# Water has 2 hydrogen atoms and 1 oxygen atom
hydrogen_count = 2
oxygen_count = 1

# Calculate contribution from each element
hydrogen_mass = hydrogen_count * atomic_masses["H"]
oxygen_mass = oxygen_count * atomic_masses["O"]

# Total molar mass
water_molar_mass = hydrogen_mass + oxygen_mass

print(f"Hydrogen contribution: {hydrogen_count} × {atomic_masses['H']} = {hydrogen_mass} g/mol")
print(f"Oxygen contribution: {oxygen_count} × {atomic_masses['O']} = {oxygen_mass} g/mol")
print(f"\n✅ Molar mass of H₂O = {round(water_molar_mass, 3)} g/mol")
```

```python
# Example 2b: Calculate molar mass of Glucose (C₆H₁₂O₆)
print("=== Calculating Molar Mass of Glucose (C₆H₁₂O₆) ===")
print("(Common in pharmaceutical formulations)\n")

# Glucose composition
carbon_count = 6
hydrogen_count = 12
oxygen_count = 6

# Calculate each element's contribution
carbon_mass = carbon_count * atomic_masses["C"]
hydrogen_mass = hydrogen_count * atomic_masses["H"]
oxygen_mass = oxygen_count * atomic_masses["O"]

# Total molar mass
glucose_molar_mass = carbon_mass + hydrogen_mass + oxygen_mass

print(f"Carbon: {carbon_count} × {atomic_masses['C']} = {carbon_mass} g/mol")
print(f"Hydrogen: {hydrogen_count} × {atomic_masses['H']} = {hydrogen_mass} g/mol")
print(f"Oxygen: {oxygen_count} × {atomic_masses['O']} = {oxygen_mass} g/mol")
print(f"\n✅ Molar mass of C₆H₁₂O₆ = {round(glucose_molar_mass, 3)} g/mol")
```

```python
# Example 2c: Simple Interactive Molar Mass Calculator
# Calculate molar mass for common compounds

def calculate_molar_mass_simple(formula_dict):
    """
    Calculate molar mass from a dictionary of elements and their counts.
    Example: {"H": 2, "O": 1} for H₂O
    """
    total_mass = 0
    
    for element, count in formula_dict.items():
        if element in atomic_masses:
            mass_contribution = count * atomic_masses[element]
            total_mass += mass_contribution
            print(f"{element}: {count} × {atomic_masses[element]} = {mass_contribution} g/mol")
        else:
            print(f"Warning: {element} not in database")
    
    return total_mass

# Test with Sulfuric Acid (H₂SO₄) - used in battery manufacturing in Uganda
print("=== Sulfuric Acid (H₂SO₄) ===")
sulfuric_acid = {"H": 2, "S": 1, "O": 4}
h2so4_mass = calculate_molar_mass_simple(sulfuric_acid)
print(f"\n✅ Molar mass of H₂SO₄ = {round(h2so4_mass, 3)} g/mol\n")

# Test with Sodium Chloride (NaCl) - table salt, food industry
print("=== Sodium Chloride (NaCl) ===")
table_salt = {"Na": 1, "Cl": 1}
nacl_mass = calculate_molar_mass_simple(table_salt)
print(f"\n✅ Molar mass of NaCl = {round(nacl_mass, 3)} g/mol")
```

### 🎯 New Python Concepts:
- **Dictionaries**: Storing key-value pairs (like a reference table)
- **Functions**: Creating reusable code with `def`
- **Loops**: Repeating actions with `for`
- **Comments**: Using `"""triple quotes"""` for documentation

## Example 3: Dilution Calculations (C₁V₁ = C₂V₂)

### The Chemistry:
When diluting solutions, the amount of solute remains constant:

**C₁V₁ = C₂V₂**

Where:
- C₁ = initial concentration
- V₁ = initial volume
- C₂ = final concentration
- V₂ = final volume

### Real-World Context:
Laboratory technicians at NWSC and pharmaceutical companies perform dilutions daily for analysis and formulation.

```python
# Example 3a: Diluting a stock solution for water quality testing
print("=== NWSC Laboratory: Preparing Standard Solutions ===")
print("Scenario: Diluting nitrate stock solution for calibration\n")

# Given values
C1 = 1000  # Stock concentration (mg/L)
C2 = 50    # Desired final concentration (mg/L)
V2 = 100   # Desired final volume (mL)

# Calculate required volume of stock solution using C1V1 = C2V2
# Rearrange: V1 = (C2 × V2) / C1
V1 = (C2 * V2) / C1

print(f"Stock concentration (C₁): {C1} mg/L")
print(f"Desired concentration (C₂): {C2} mg/L")
print(f"Desired final volume (V₂): {V2} mL")
print(f"\n✅ Required stock volume (V₁): {V1} mL")

# Calculate amount of water to add
water_volume = V2 - V1
print(f"\n📋 Lab Protocol:")
print(f"1. Pipette {V1} mL of stock solution")
print(f"2. Add {water_volume} mL of distilled water")
print(f"3. Mix thoroughly")
print(f"4. Final volume: {V2} mL at {C2} mg/L")
```

```python
# Example 3b: Interactive Dilution Calculator
print("=== Interactive Dilution Calculator ===")
print("Calculate any unknown value in C₁V₁ = C₂V₂\n")

def dilution_calculator(C1=None, V1=None, C2=None, V2=None):
    """
    Dilution calculator using C1V1 = C2V2
    Provide 3 values, and it calculates the 4th
    """
    
    # Count how many values are provided
    values = [C1, V1, C2, V2]
    none_count = values.count(None)
    
    if none_count != 1:
        print("Error: Provide exactly 3 values and leave 1 as None")
        return
    
    # Calculate the missing value
    if C1 is None:
        C1 = (C2 * V2) / V1
        print(f"✅ Calculated C₁ = {round(C1, 2)}")
    elif V1 is None:
        V1 = (C2 * V2) / C1
        print(f"✅ Calculated V₁ = {round(V1, 2)}")
    elif C2 is None:
        C2 = (C1 * V1) / V2
        print(f"✅ Calculated C₂ = {round(C2, 2)}")
    elif V2 is None:
        V2 = (C1 * V1) / C2
        print(f"✅ Calculated V₂ = {round(V2, 2)}")
    
    # Verify the equation
    left_side = C1 * V1
    right_side = C2 * V2
    print(f"\nVerification: C₁V₁ = {round(left_side, 2)}, C₂V₂ = {round(right_side, 2)}")
    print(f"Match: {abs(left_side - right_side) < 0.01} ✅")

# Example 1: Find required stock volume
print("\n--- Example 1: What volume of stock is needed? ---")
dilution_calculator(C1=500, V1=None, C2=25, V2=200)

# Example 2: Find final concentration
print("\n--- Example 2: What's the final concentration? ---")
dilution_calculator(C1=100, V1=10, C2=None, V2=250)
```

```python
# Example 3c: Serial Dilution Calculator (Advanced)
print("=== Serial Dilution for Quality Control ===")
print("Common in pharmaceutical and water testing labs\n")

# Starting with 1000 mg/L stock, create 5 standards
initial_concentration = 1000  # mg/L
dilution_factor = 2  # Each dilution is half the previous
number_of_standards = 5

print(f"Starting concentration: {initial_concentration} mg/L")
print(f"Dilution factor: 1:{dilution_factor}\n")
print("Standard Series:")
print("-" * 30)

current_concentration = initial_concentration

for i in range(number_of_standards):
    standard_number = i + 1
    print(f"Standard {standard_number}: {current_concentration} mg/L")
    
    # Calculate next concentration
    current_concentration = current_concentration / dilution_factor

print("\n✅ These standards can be used to create a calibration curve!")
```
